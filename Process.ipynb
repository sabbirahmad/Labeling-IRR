```python
def getjimmydata(file):
    with open(file, 'r') as json_file:  
#         old_res = []
#         for line in json_file:
#             if line[-2] == ',':
#                 line = line[:-2]
#             data = eval(line)
#             old_res.append(data)
            
        old_res = json.load(json_file)
    return old_res

def getericadata(file):
    with open(file, 'r') as json_file:  
#         new_res = []
#         for line in json_file:
#             if line[-2] == ',':
#                 line = line[:-2]
#             data = eval(line)
#             new_res.append(data)
        new_res = json.load(json_file)
            
    return new_res

def overlap(label_1, label_2):
    if label_1['timestamp_range'][0] >= label_2['timestamp_range'][0] and label_1['timestamp_range'][0] < label_2['timestamp_range'][1]:
        if label_1['units'] == label_2['units']:
            return True
    if label_1['timestamp_range'][1] <= label_2['timestamp_range'][1] and label_1['timestamp_range'][1] > label_2['timestamp_range'][0]:
        if label_1['units'] == label_2['units']:
            return True
    if label_1['timestamp_range'][0] < label_2['timestamp_range'][0] and label_1['timestamp_range'][1] > label_2['timestamp_range'][1]:
        if label_1['units'] == label_2['units']:
            return True
    return False

def createNextRow(df, jimmy_label, erica_label):
    new_row = {}
    
    if jimmy_label['title'] == 'Explosive Quick Mine':
        new_row['Jimmy Explosive Quick Mine'] = 1
    elif jimmy_label['title'] == 'Explosive Tunneling':
        new_row['Jimmy Explosive Tunneling'] = 1
    elif jimmy_label['title'] == 'Explosive Inner':
        new_row['Jimmy Explosive Inner'] = 1
    elif jimmy_label['title'] == 'Exploring':
        new_row['Jimmy Exploring'] = 1
    elif jimmy_label['title'] == 'Tunneling Pickaxe':
        new_row['Jimmy Tunneling Pickaxe'] = 1
    elif jimmy_label['title'] == 'Pickaxing for Gold':
        new_row['Jimmy Pickaxing for Gold'] = 1
    elif jimmy_label['title'] == 'Explosive Miss':
        new_row['Jimmy Explosive Miss'] = 1    
    
    new_row['Jimmy TimeStamp'] = jimmy_label['timestamp_range']
    
    if erica_label['title'] == 'Explosive Quick Mine':
        new_row['Erica Explosive Quick Mine'] = 1
    elif erica_label['title'] == 'Explosive Tunneling':
        new_row['Erica Explosive Tunneling'] = 1
    elif erica_label['title'] == 'Explosive Inner':
        new_row['Erica Explosive Inner'] = 1
    elif erica_label['title'] == 'Exploring':
        new_row['Erica Exploring'] = 1
    elif erica_label['title'] == 'Tunneling Pickaxe':
        new_row['Erica Tunneling Pickaxe'] = 1
    elif erica_label['title'] == 'Pickaxing for Gold':
        new_row['Erica Pickaxing for Gold'] = 1
    elif erica_label['title'] == 'Explosive Miss':
        new_row['Erica Explosive Miss'] = 1   
    
    new_row['Erica TimeStamp'] = erica_label['timestamp_range']
    return new_row

def ericanolabel(jimmy_label):
    new_row={}
    
    if jimmy_label['title'] == 'Explosive Quick Mine':
        new_row['Jimmy Explosive Quick Mine'] = 1
    elif jimmy_label['title'] == 'Explosive Tunneling':
        new_row['Jimmy Explosive Tunneling'] = 1
    elif jimmy_label['title'] == 'Explosive Inner':
        new_row['Jimmy Explosive Inner'] = 1
    elif jimmy_label['title'] == 'Exploring':
        new_row['Jimmy Exploring'] = 1
    elif jimmy_label['title'] == 'Tunneling Pickaxe':
        new_row['Jimmy Tunneling Pickaxe'] = 1
    elif jimmy_label['title'] == 'Pickaxing for Gold':
        new_row['Jimmy Pickaxing for Gold'] = 1
    elif jimmy_label['title'] == 'Explosive Miss':
        new_row['Jimmy Explosive Miss'] = 1  
    
    new_row['Jimmy TimeStamp'] = jimmy_label['timestamp_range']
    new_row['Erica No label'] = 1
    
    return new_row

def jimmynolabel(erica_label):
    new_row={}
    
    if erica_label['title'] == 'Explosive Quick Mine':
        new_row['Erica Explosive Quick Mine'] = 1
    elif erica_label['title'] == 'Explosive Tunneling':
        new_row['Erica Explosive Tunneling'] = 1
    elif erica_label['title'] == 'Explosive Inner':
        new_row['Erica Explosive Inner'] = 1
    elif erica_label['title'] == 'Exploring':
        new_row['Erica Exploring'] = 1
    elif erica_label['title'] == 'Tunneling Pickaxe':
        new_row['Erica Tunneling Pickaxe'] = 1
    elif erica_label['title'] == 'Pickaxing for Gold':
        new_row['Erica Pickaxing for Gold'] = 1
    elif erica_label['title'] == 'Explosive Miss':
        new_row['Erica Explosive Miss'] = 1    
    new_row['Erica TimeStamp'] = erica_label['timestamp_range']
    
    new_row['Jimmy No label'] = 1
    
    return new_row


test = getericadata('./data/boomtown/original/8_11_2018__3_37_36_0.json')
print(len(test))
```
